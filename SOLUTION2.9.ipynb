```python
# 2. 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week


query1 = '''SELECT *, 
CASE
WHEN weekday(rental_date)< 6 THEN 'workday'
ELSE 'weekend'
END AS 'day_type'
FROM sakila.rental;'''
day= pd.read_sql_query (query1, engine)
day.head()
```

```python
# 3. Get all films with ARMAGEDDON in the title

query2="""select* from sakila.film
where title='ARMAGEDDON'"""
day= pd.read_sql_query (query2, engine)
day.head()
```

```python
# 4. Get 10 the longest films.
query3="""SELECT title, length  from sakila.film
order by length desc
limit 10"""
day= pd.read_sql_query (query3, engine)
day.head(10)

```

```python
# 5.  How many films include Behind the Scenes content?

query4="""select* from sakila.film
where special_features='Behind the Scenes'"""
day= pd.read_sql_query (query4, engine)
day.head()
```

```python
#  6. Which kind of movies (rating) have a mean duration of more than two hours?
query5="""select rating, avg(length/60) from sakila.film group by rating
having  avg(length/60) > 2;"""
day= pd.read_sql_query (query5, engine)
day.head()
```

```python
# 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.*/

query6="""select length, title from sakila.film
order by length desc;"""
day= pd.read_sql_query (query6, engine)
day.head()
```
